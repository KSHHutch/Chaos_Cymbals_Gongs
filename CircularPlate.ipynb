# Vibrational Displacement on a Circular Plate

```python
from scipy.special import jv, iv
from numpy import cos

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
mpl.rcParams['figure.figsize'] = (10,10)
```

```python
# Create mesh in polar coordinates and compute displacement Y
r = np.linspace(0, 1, 500)
theta = np.linspace(0, 2*np.pi, 500)
R, Theta = np.meshgrid(r, theta)
n = 6
A = 1
B = 1
Y = ( A*jv(n,R) + B*iv(n,R) ) * cos(n*Theta)

# Express the mesh in the cartesian system.
X, Z = R*np.cos(Theta), R*np.sin(Theta)

# Instantiate figure
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Plot the surface.
ax.plot_surface(X, Z, Y, cmap=plt.cm.YlGnBu_r)

# Tweak the limits
ax.set_zlim(-0.0003, 0.0003)
ax.set_xlabel('x')
ax.set_ylabel('z')
ax.set_zlabel('y')
ax.set_title(f'Circular plate mode n={n}')
plt.rc('font', size=13) 
```
